```python
from pyfoma import *

Grammar = {}
Grammar["S"] = [("", "Verb")]
Grammar["Verb"] = [("vivir", "V"), ("comer","V"), ("bailar", "V"), ("hablar", "V"), ("correr", "V"), ("escribir", "V")]
                   #, ("hablar","V"),\
                   #("correr","V"),("escribir","V")]

#("hablar", "V"), 

Grammar["V"] = [(("'[V]'", ""), "VInfl")]
                

#all the endings are for -ar verbs, I am going to base my rewrite rules off of this
Grammar["VInfl"] = [(("'[1p]' '[Sg]' '[Pres]'", "+o"), "#"), \
                    (("'[2p]' '[Sg]' '[Pres]'", "+as"), "#"), \
                    (("'[3p]' '[Sg]' '[Pres]'", "+a"), "#"), \
                   ((" '[1p]' '[Pl]' '[Pres]'", "+amos"), "#"),\
                   (("'[2p]' '[Pl]' '[Pres]'", "+áis"), "#"),\
                   (("'[3p]' '[Pl]' '[Pres]'", "+an"), "#"),\
                (("'[1p]' '[Sg]''[Past]'", "+é"), "#"), \
                    (("'[2p]' '[Sg]' '[Past]'", "+aste"), "#"), \
                    (("'[3p]' '[Sg]''[Past]'", "+ó"), "#"), \
                   ((" '[1p]' '[Pl]''[Past]'", "+amos"), "#"),\
                   (("'[2p]' '[Pl]''[Past]' ", "+asteis"), "#"),\
                   (("'[3p]' '[Pl]' '[Past]'", "+aron"), "#"),\
                   (("'[1p]' '[Sg]' '[Subj]'", "+e"), "#"), \
                    (("'[2p]' '[Sg]' '[Subj]'", "+es"), "#"), \
                    (("'[3p]' '[Sg]' '[Subj]'", "+e"), "#"), \
                   ((" '[1p]' '[Pl]' '[Subj]'", "+emos"), "#"),\
                   (("'[2p]' '[Pl]' '[Subj]'", "+éis"), "#"),\
                   (("'[3p]' '[Pl]' '[Subj]'", "+en"), "#")]


#got these few lines from HW2
Lexicon = FST.rlg(Grammar, "S")
Lexicon = Lexicon.epsilon_remove().determinize().minimize() 
Lexicon.view()
print(Paradigm(Lexicon, ".*"))
```

```python
fsts = {} # Let's store all our FSTs in this dictionary
fsts['Lexicon'] = Lexicon

#2p present for -ir
fsts['rule1'] = FST.re("$^rewrite(((ir'+'as):(ires)) /  _'')")

#2- present for -er
fsts['rule2'] = FST.re("$^rewrite((er'+'as):(eres) / _ '')")

#3p present for er and ir
fsts['rule3'] = FST.re("$^rewrite((er'+'a):(ere))")

fsts['rule4'] = FST.re("$^rewrite((ir'+'a):(ire))")

#1p plural present and pretertie for -ir
fsts['rule5'] = FST.re("$^rewrite((iremos):(irimos))")
#its iremos instead of iramos cuz i thanges ir+a to ire

#1p plural present for -er
fsts['rule6'] = FST.re("$^rewrite((eremos):(er'+'imos))")

#[2p][Pl][Pres] for er
fsts['rule7'] = FST.re("$^rewrite((er'+'áis):(eréis))")

#[2p][Pl][Pres] for ir
fsts['rule8'] = FST.re("$^rewrite(((ir'+'áis):(ir'+'ís)))")

#2p plural pres for er
fsts['rule9'] = FST.re("$^rewrite((er'+'an):(er'+'en))")

#2p plural pres for ir
fsts['rule10'] = FST.re("$^rewrite((ir'+'an):(ir'+'en))")

#i removed the plus in many of the rewrite rules so when I started to do other tenses it wouldn't be rewriting the previous rules because
#many of the conjugations include the same letters, so the '+' allows me to change specific endings

#1p sing pret for ir and er
fsts['rule11'] = FST.re("$^rewrite((é:í) /ir'+'_'')")

#2p plural present for -er
fsts['rule12'] = FST.re("$^rewrite((er'+'éis):(er'+'áis))")

#2nd sing and plur pret for ir and er
fsts['rule13'] = FST.re("$^rewrite((este):('+'iste) /er|ir_'')")

#3rd sing pret for ir and er
fsts['rule14'] =  FST.re("$^rewrite((ó:'ió') /er'+'|ir'+'_'')")

#3p plural pret for er and ir
fsts['rule15'] = FST.re("$^rewrite((eron):('+'ieron) /er|ir_'')")

#1p and 3p sing for subj
fsts['rule16'] = FST.re("$^rewrite(('+'e):(a)/er|ir_'' )")

#2p plural present for -ir
fsts['rule17'] = FST.re("$^rewrite((íis):(áis))")
#is it íis instead of éis cuz i changed é to í in rule11

#removing verb endings
fsts['rule18'] = FST.re("$^rewrite((((er)|(ar)|(ir)):('')))")

#3p plural past for er and ir verbs
fsts['rule19'] = FST.re("$^rewrite(('+'ion):(ieron))")

#3p plural past for ar verbs
fsts['rule20'] = FST.re("$^rewrite(('+'on):(aron))")

#rule 19 and 20 come after I remove the verb endings because the conjucations include the same letters as the verb endings, so I need to put the letters back
#in the conjugations

fsts['cleanup'] = FST.re("$^rewrite('+':'')")

grammar = FST.re("$Lexicon @$rule1 @$rule2 @$rule3 @$rule4 @$rule5 @$rule6 @$rule7 @$rule8 @$rule9 @$rule10 @$rule11 \
@$rule12 @$rule13 @$rule14 @$rule15 @$rule16 @$rule17 @$rule18 @$rule19 @$rule20 @$cleanup",fsts)
                
print(Paradigm(grammar, ".*"))

#er verbs 1p plural past is incorrect.
#past 1p plural and pres 1p plural have the same endings in -ar and -ir, -er does not. I couldn't find a way to
#change one without changing the other in -er

#Once I addeed the Subjunctive tense, it got a lot more difficult to make my code concise/short.
```
